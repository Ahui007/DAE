```python
import numpy as np
import pandas as pd
from keras.layers import Dense,Dropout
import matplotlib.pyplot as plt
from keras.optimizers import Adam,rmsprop
from keras.models import Sequential,Model
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn import metrics
from random import sample
import time
def purity_score(y_true, y_pred):
    # compute contingency matrix (also called confusion matrix)
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    # return purity
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)
```

```python
expr = pd.read_csv('bench1_1000genes.csv',index_col=0)
```

```python
X = expr.values[:,:-1]
Y = expr.values[:,-1]
Y = pd.factorize(Y)[0]
K = len(np.unique(Y))
print('{} different cell types'.format(K))
ncell = X.shape[0]
ngene = X.shape[1]
X = X.astype('float')
print('proportion of 0s: {}'.format(np.mean(X==0)))
```

```python
libs = X.sum(axis = 1)
norm_factor = np.diag(np.median(libs)/libs)
X_norm = np.dot(norm_factor,X)
X_norm = np.log2(X_norm + 1.)
X_norm = StandardScaler().fit_transform(X_norm)
```

```python
pca = PCA(n_components = 50, random_state=1)
PCA_raw = pca.fit_transform(X_norm)
tsne = TSNE(n_components = 2,random_state=1)
T_sne_raw = tsne.fit_transform(PCA_raw)
```

```python
unique_class = np.unique(Y)
for i in range(K):
    plt.scatter(T_sne_raw[Y==unique_class[i],0],
                T_sne_raw[Y==unique_class[i],1], 
                s=5,label = unique_class[i] )
plt.title('T-sne plot for raw data')
```

#### convergence study

```python
#del model1
t = time.time()
n_input = X.shape[1]
prob_mat = np.exp(-1.72794*(np.log(X+1)-0.65366))
prob_mat[prob_mat>1]=1 #dropout probability

zero_mask = np.random.binomial(1,1-prob_mat,[ncell,ngene])
X_noise = X.copy()
X_noise = X_noise*zero_mask #add dropout to the observed data

#X_noise = np.exp(X_noise)####

model1 = Sequential()
model1.add(Dense(128,activation = 'relu',
            input_shape = (n_input,)))

model1.add(Dense(64,activation = 'relu',
            name = 'bottleneck'))

model1.add(Dense(128,activation='relu'))
model1.add(Dense(n_input,activation = 'relu'))
#train test split
training_sample = sample(range(ncell),150)
testing_sample = list(set(np.arange(ncell))-set(training_sample))
#model training
model1.compile(loss='mean_squared_error',
                optimizer=rmsprop(learning_rate=0.001))
history = model1.fit(X_noise[training_sample,:],X[training_sample,:],
                     batch_size=32,epochs=1000,verbose=0,shuffle=True,
                    validation_data=(X_noise[testing_sample,:],X[testing_sample,:]))
Denoise = Model(model1.input,model1.output)
X_imp = Denoise.predict(X)
print(time.time()-t)
```

```python
plt.plot(history.history['loss'],label = 'loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
print(history.history['loss'][-1],history.history['val_loss'][-1])
```

#### test model

```python
t = time.time()
X_imp = np.zeros(shape=(ncell,ngene))
runs = 50
n_input = X.shape[1]
for i in range(runs):
    prob_mat = np.exp(-1.72794*(np.log(X+1)-0.65366)) #dropout probability
    prob_mat[prob_mat>1] = 1
    zero_mask = np.random.binomial(1,1-prob_mat,[ncell,ngene])
    X_noise = X.copy()
    X_noise = X_noise*zero_mask #add dropout to the observed data
    
    #build autoencoder model
    
    model1 = Sequential()
    model1.add(Dense(128,activation = 'relu',
                input_shape = (n_input,)))
    #model1.add(Dense(64,activation='relu'))
    model1.add(Dense(64,activation = 'relu',
                name = 'bottleneck'))
    #model1.add(Dense(64,activation='relu'))
    model1.add(Dense(128,activation='relu'))
    model1.add(Dense(n_input,activation = 'relu'))
    #model training
    model1.compile(loss='mean_squared_error',
                   optimizer=rmsprop())
    model1.fit(X_noise,X,batch_size=32,epochs=800,
           verbose=0,shuffle=True)
    Denoise = Model(model1.input,model1.output)
    X_imp = X_imp + Denoise.predict(X)
    del model1, Denoise
X_imp = X_imp/runs
print('elapse time: {}'.format(time.time()-t)) 
```

```python

libs = X_imp.sum(axis=1)

norm_factor = np.diag(np.median(libs)/libs)
X_imp_norm = np.dot(norm_factor,X_imp)
X_imp_norm = np.log2(X_imp_norm + 1.)
for i in range(ngene): 
    if np.std(X_imp_norm[:,i])!=0:
        X_imp_norm[:,i] = (X_imp_norm[:,i]-np.mean(X_imp_norm[:,i]))/np.std(X_imp_norm[:,i])
```

```python
pca = PCA(n_components = 50, random_state=1)
PCA_imp = pca.fit_transform(X_imp_norm)
tsne = TSNE(n_components = 2,random_state=1)
T_sne_imp = tsne.fit_transform(PCA_imp)
```

```python
unique_class = np.unique(Y)
for i in range(K):
    plt.scatter(T_sne_imp[Y==unique_class[i],0],
                T_sne_imp[Y==unique_class[i],1], 
                s=5,label = unique_class[i] )
plt.title('T-sne plot for raw data')
```

```python
kmeans = KMeans(n_clusters = K,random_state = 42).fit(T_sne_imp)
print('purity score is: {}'.format(purity_score(Y,kmeans.labels_)))
print('ARI is {}'.format(metrics.adjusted_rand_score(Y,kmeans.labels_)))
```
